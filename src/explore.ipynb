```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from dotenv import load_dotenv
import requests

# load the .env file variables
load_dotenv('/content/api.env')

# Get credential values
client_id = os.environ.get("CLIENT_ID")
client_secret = os.environ.get("CLIENT_SECRET")
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
```

```python
uri = 'spotify:artist:4dwdTW1Lfiq0cM8nBAqIIz'

spotify = spotipy.Spotify(client_credentials_manager=auth_manager)
results = spotify.artist_top_tracks(uri)

canciones = pd.DataFrame()
i = 1

for track in results['tracks'][:10]:
    canciones[i] = {
        'nombre': track['name'],
        'popularidad': track['popularity'],
        'duracion': track['duration_ms']/60000
    }
    i += 1
```

```python
canciones = canciones.T
```
